```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

국공립유치원 클러스터링

```python
k1 = pd.read_csv('data/kinder1. capa_result.csv',encoding='CP949')
k2 = pd.read_csv('data/kinder2. dutynot_result.csv',encoding='CP949')
k3 = pd.read_csv('data/kinder3. no400m_result.csv',encoding='CP949')
k4 = pd.read_csv('data/kinder4. completeno_result.csv',encoding='CP949')
```

```python
k1=k1[['gu','dong','result']]
k2=k2[['gu','dong','ee']]
k3=k3[['gu','dong','ee']]
k4=k4[['gu','dong','ee']]
```

```python
print(len(k1),len(k2),len(k3),len(k4))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
k2[['ee']]=scaler.fit_transform(k2[['ee']])
```

```python
k1[['result']]=scaler.fit_transform(k1[['result']])
k3[['ee']]=scaler.fit_transform(k3[['ee']])
k4[['ee']]=scaler.fit_transform(k4[['ee']])
```

```python
k1=k1.rename(columns={'result':'r1'})

k2=k2.drop(columns='gu')
k2=k2.rename(columns={'ee':'r2'})

k3=k3.drop(columns='gu')
k3=k3.rename(columns={'ee':'r3'})

k4=k4.drop(columns='gu')
k4=k4.rename(columns={'ee':'r4'})
```

```python
k1=pd.merge(k1,k2,how='left',on='dong')
k1=pd.merge(k1,k3,how='left',on='dong')
k1=pd.merge(k1,k4,how='left',on='dong')
k1
```

```python
data=k1.values[:,2:]
data
```

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
Sum_of_squared_distances
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans=KMeans(n_clusters=7).fit(data)
```

```python
kmeans.labels_
```

```python
k1['cluster_id']=kmeans.labels_
```

```python
k1
```

```python
k1.to_excel(r"C:\Users\user\Desktop\molra\kindercluster.xlsx")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data)
```

```python
# Visualize two scatter
# Your Code Here
X_pca = data_pca[:,0]
Y_pca = data_pca[:,1]
# End Your Code

colors = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0']


# Your Code Here
C_kmeans = kmeans.labels_
# End Your Code

for i in range(7):
    px = X_pca[C_kmeans == i]
    py = Y_pca[C_kmeans == i]
    plt.scatter(px,py, c=colors[i], s=5, label=i)
    
plt.legend(loc='upper right')
plt.title("K-means Clustering")
plt.show()
```

국공립어린이집 클러스터링

```python
p1 = pd.read_csv('data/pre1. capa_result.csv',encoding='CP949')
p2 = pd.read_csv('data/pre2. dutynot_result.csv',encoding='CP949')
p3 = pd.read_csv('data/pre3. no400m_result.csv',encoding='CP949')
p4 = pd.read_csv('data/pre4. completeno_result.csv',encoding='CP949')
```

```python
p1=p1[['gu','dong','result']]
p2=p2[['gu','dong','ee']]
p3=p3[['gu','dong','ee']]
p4=p4[['gu','dong','ee']]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
p1[['result']]=scaler.fit_transform(p1[['result']])
p2[['ee']]=scaler.fit_transform(p2[['ee']])
p3[['ee']]=scaler.fit_transform(p3[['ee']])
p4[['ee']]=scaler.fit_transform(p4[['ee']])
```

```python
p1=p1.rename(columns={'result':'r1'})

p2=p2.drop(columns='gu')
p2=p2.rename(columns={'ee':'r2'})

p3=p3.drop(columns='gu')
p3=p3.rename(columns={'ee':'r3'})

p4=p4.drop(columns='gu')
p4=p4.rename(columns={'ee':'r4'})

p1=pd.merge(p1,p2,how='left',on='dong')
p1=pd.merge(p1,p3,how='left',on='dong')
p1=pd.merge(p1,p4,how='left',on='dong')
p1
```

```python
pdata=p1.values[:,2:]

Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(pdata)
    Sum_of_squared_distances.append(km.inertia_)
Sum_of_squared_distances
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans=KMeans(n_clusters=6).fit(pdata)
```

```python
p1['cluster_id']=kmeans.labels_
p1
```
